```python
# Program to display even numbers from 1 to 10

for i in range(1, 11):
    if i % 2 == 0:
        print(i)
```

```python
#Program to add odd numbers from 1 to 10

sum = 0

for i in range(1, 11):
    if i % 2 != 0:
        sum = sum + i

print("Sum of odd numbers =", sum)
```

```python
#Python program to print Fibonacci series between 0 and 50

a = 0
b = 1

print("Fibonacci series between 0 and 50:")

while a <= 50:
    print(a)
    c = a + b
    a = b
    b = c
```
