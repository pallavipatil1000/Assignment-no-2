```python
# Create a list
numbers = [10, 20, 30, 40, 50]

# Check membership
print(numbers)

# Using 'in'
print(20 in numbers)
print(60 in numbers)

# Using 'not in'
print(60 not in numbers)
print(10 not in numbers)
```

```python
# Create a list
fruits = ["apple", "banana", "cherry", "mango", "grapes"]

print(fruits)

# Indexing
print("First element:", fruits[0])
print("Third element:", fruits[2])

# Negative indexing
print("Last element:", fruits[-1])
print("Second last:", fruits[-2])

# Slicing
print("Elements from index 1 to 3:", fruits[1:4])
print("First three elements:", fruits[:3])
print("Last two elements:", fruits[-2:])
```

```python
# Create a list
numbers = [1, 2, 3, 4, 5]

print(numbers)

# Update
numbers[2] = 10
print("After update:", numbers)

# Append
numbers.append(6)
print("After append:", numbers)

# Insert
numbers.insert(1, 20)
print("After insert:", numbers)

# Delete (remove by value)
numbers.remove(4)
print("After remove:", numbers)

# Delete (pop by index)
numbers.pop(2)
print("After pop:", numbers)

# Delete entire list element
del numbers[0]
print("After del:", numbers)
```

```python
# Create lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Concatenation
result = list1 + list2
print("Concatenation:", result)

# Repetition
print("Repetition:", list1 * 2)

# Length
print("Length of list1:", len(list1))

# Maximum and Minimum
print("Maximum:", max(list2))
print("Minimum:", min(list2))
```
